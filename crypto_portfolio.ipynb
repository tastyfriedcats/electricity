```python
import gdax
import pandas as pd
import numpy as np
from datetime import datetime
import datetime
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from matplotlib import pylab
import plotly.plotly as py
import plotly.graph_objs as go
from dateutil import parser
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 12
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size
print "Current size:", fig_size
```

```python
client = gdax.PublicClient()

ETH = client.get_product_historic_rates('ETH-USD', granularity = 60 * 60 * 24)
eth = pd.DataFrame(ETH)

BTC = client.get_product_historic_rates('BTC-USD', granularity = 60 * 60 * 24)
btc = pd.DataFrame(BTC)

LTC = client.get_product_historic_rates('LTC-USD', granularity = 60 * 60 * 24)
ltc = pd.DataFrame(LTC)
```

```python
for i in range(len(ETH)):
    eth[0][i] = (
    datetime.datetime.fromtimestamp(
        int(eth[0][i])
    ).strftime('%Y-%m-%d %H:%M:%S')
    )

for i in range(len(BTC)):
    btc[0][i] = (
    datetime.datetime.fromtimestamp(
        int(btc[0][i])
    ).strftime('%Y-%m-%d %H:%M:%S')
    )
    
for i in range(len(LTC)):
    ltc[0][i] = (
    datetime.datetime.fromtimestamp(
        int(ltc[0][i])
    ).strftime('%Y-%m-%d %H:%M:%S')
    )
```

```python
for i in range(len(ETH)):
    eth[0][i] = parser.parse(eth[0][i])

for i in range(len(BTC)):
    btc[0][i] = parser.parse(btc[0][i])

for i in range(len(LTC)):
    ltc[0][i] = parser.parse(ltc[0][i])
```

```python
eth.columns = ['Time', 'ETH Low', 'ETH High', 'ETH Open', 'ETH Close', 'ETH Volume']
eth = eth.iloc[::-1]
eth = eth.reset_index(drop = True)

btc.columns = ['BTC Time', 'BTC Low', 'BTC High', 'BTC Open', 'BTC Close', 'BTC Volume']
btc = btc.iloc[::-1]
btc = btc.reset_index(drop = True)

ltc.columns = ['LTC Time', 'LTC Low', 'LTC High', 'LTC Open', 'LTC Close', 'LTC Volume']
ltc = ltc.iloc[::-1]
ltc = ltc.reset_index(drop = True)
```

```python
df = pd.concat([eth, btc, ltc], axis = 1)
df.drop('BTC Time', 1, inplace = True)
df.drop('LTC Time', 1, inplace = True)
df
```

```python
eth_return = np.diff(np.log(eth['Close']))
n = len(eth_return)
mean_eth_return = eth_return.mean()
eth_std = eth_return.std()

btc_return = np.diff(np.log(btc['Close']))
n = len(btc_return)
mean_btc_return = btc_return.mean()
btc_std = btc_return.std()

ltc_return = np.diff(np.log(ltc['Close']))
n = len(ltc_return)
mean_ltc_return = ltc_return.mean()
ltc_std = ltc_return.std()

print ('The mean return of eth is {}, ' 
        'and the volatility is {}').format(mean_eth_return, eth_std)
```
